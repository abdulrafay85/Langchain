### Text Splitters Kya Hain?

**Text Splitters** wo tools hain jo aapki lambi text ko chhote chhote hisson mein baantne ka kaam karte hain. Ye zaroori hota hai kyunke jab aap lambi documents ya text ko computer ya AI model ko dete hain, to wo zyada samajhne mein mushkil hoti hain. 

### Kyu Zaroori Hain?

1. **Lambi Text Ko Manage Karna**: Agar aapke paas bohot lambi text hai, to usse chhote hisson mein baantna asaan hota hai. Jaise agar aapke paas ek bohot lambi kahani hai, to usse chhote paragraphs ya sentences mein baantne se aapko samajhna asaan hota hai.

2. **Context Rakhna**: Jab aap text ko baantte hain, to kuch hisson ko ek doosre se juda rakhte hain taake unka matlab ya context khatam na ho. Jaise, agar aap ek paragraph se do sentence nikaal rahe hain, to wo dono sentences ek doosre se juda rahna chahiye.

### Kaise Kaam Karte Hain?

1. **Chhote Hisson Mein Baatna**: Pehle, text ko chhote hisson (jaise sentences) mein baanta jata hai. Ye aise hota hai ke jab aap ek lambi kahani padte hain, to uske chhote chhote hisse aasan hota hai samajhna.

2. **Bade Hisse Banana**: Phir, ye chhote hisse ko milakar ek bada hissa banate hain jab tak wo ek certain size tak nahi pahunchte. Jaise aap chhote paragraphs bana rahe hain.

3. **Overlap Rakhna**: Jab ek bada hissa banate hain, to kabhi kabhi pehle wale hisson ke kuch hisson ko naya hissa banate waqt istemal karte hain. Ye isliye hota hai taake jab aap naya hissa shuru karte hain, to pehle wale ka context yaani maqsad samajh aaye.

### Kya Customize Kar Sakte Hain?

- **Kaise Baantna Hai**: Aap decide kar sakte hain ke text ko kaise baantna hai. Kya aap har sentence mein baantna chahte hain ya paragraphs mein?

- **Chunk Size**: Aap yeh bhi decide kar sakte hain ke har hissa kitna bada hona chahiye. Kya wo 50 words ka hona chahiye ya 100 words ka?

### Summary

Text Splitters lambi text ko chhote, samajhne mein asaan hisson mein baantte hain. Ye aapko text ko achhe se manage karne aur samajhne mein madad karte hain.

### **character splitting**  
 **ye guarantee nahi karta hay. ke jo ap is ko chunk size dengay os hi chunk size me ap ka data mile ga.**

**character splitting** ke concept ko explain karta hai, jo ek simple tareeqa hai text ko chhote chhote pieces (chunks) mein todne ka, based on the number of characters. Let’s go step by step samajhte hain:

### Character Splitting Ka Concept:
Character splitting ka matlab yeh hai ke aap apne text ko **fixed number of characters** ke hisaab se todte ho, bina dekhe ke wo sentence complete hai ya nahi. Yeh splitting kisi bhi structure ka khayal nahi karti, bas text ko n-characters ke chunks mein divide karti hai.

**Pros (Fayde)**:
- Simple aur asaan tareeqa hai.
- Har chunk ki size ko easily control kar sakte ho.

**Cons (Nuqsanat)**:
- Yeh method rigid hota hai, yani ke yeh text ka structure ya meaning ka khayal nahi rakhta.
- Yeh sentences ko beech mein tod sakta hai, jo context ko samajhne mein mushkilat paida karta hai.

### Important Concepts:
1. **Chunk Size**: 
   - Chunk size se murad yeh hai ke aap kitne characters ek chunk mein chahte ho. 
   - For example, agar aap chunk size `35` set karte ho, to har chunk 35 characters ka hoga.
   
2. **Chunk Overlap**:
   - Overlap ka matlab yeh hai ke ek chunk ka thoda hissa next chunk mein bhi repeat ho. Is se context loss hone ka chance kam hota hai.
   - For example, agar aap overlap `4` set karte ho, to pehla chunk ke last ke 4 characters agle chunk ke shuru ke 4 characters banenge. Yeh duplication help karta hai taake context miss na ho.

### Code Example:

Let’s explain the example step by step:

1. **Sample Text**:
   ```python
   text = "This is the text I would like to chunk up. It is the example text for this exercise"
   ```

   Yeh text hai jo humne split karna hai.

2. **Manual Splitting**:
   Yahan pe manually text ko split kiya gaya hai.

   ```python
   chunks = []
   chunk_size = 35  # Chunk size is 35 characters

   # Loop through the text and split into 35-character chunks
   for i in range(0, len(text), chunk_size):
       chunk = text[i:i + chunk_size]
       chunks.append(chunk)

   chunks
   ```

   Is code mein `chunk_size = 35` diya hai, yani ke har chunk 35 characters ka hoga. Phir loop chalake text ko is size ke hisaab se tod diya gaya hai.

   Result:
   ```python
   ['This is the text I would like to ch',
    'unk up. It is the example text for ',
    'this exercise']
   ```

3. **CharacterTextSplitter Use Karna**:
   Ab yeh manual splitting karne ke bajaye, hum **LangChain** ka `CharacterTextSplitter` use karenge jo text ko chunks mein automatically tod deta hai:

   ```python
   from langchain.text_splitter import CharacterTextSplitter

   text_splitter = CharacterTextSplitter(
       chunk_size = 35,       # Chunk size
       chunk_overlap=0,       # No overlap initially
       separator='',          # Separator koi nahi
       strip_whitespace=False # White spaces ko remove nahi karna
   )

   text_splitter.create_documents([text])
   ```

   Output:
   ```python
   [Document(page_content='This is the text I would like to ch'),
    Document(page_content='unk up. It is the example text for '),
    Document(page_content='this exercise')]
   ```

   Yahan pe text ko 35 characters ke chunks mein tod diya gaya hai aur yeh chunks **Document** objects mein store ho rahe hain, jo baad mein filtering aur processing mein kaam aayenge.

4. **Chunk Overlap**:
   Overlap ka fayda yeh hota hai ke har chunk ke end ka kuch hissa next chunk ke start mein aa jata hai. Is se context ka loss nahi hota. Example:

   ```python
   text_splitter = CharacterTextSplitter(
       chunk_size = 35,
       chunk_overlap=4,   # 4 character overlap
       separator=''
   )
   text_splitter.create_documents([text])
   ```

   Output:
   ```python
   [Document(page_content='This is the text I would like to ch'),
    Document(page_content='o chunk up. It is the example text'),
    Document(page_content='ext for this exercise')]
   ```

   Yahan dekho ke first chunk ke last ke `o ch` next chunk ke shuru mein bhi aa gaye hain. Yeh overlapping se ensure hota hai ke koi important context lose na ho.

### Summary:
- **Character Splitting**: Text ko fixed number of characters ke hisaab se split karna.
- **Chunk Overlap**: Ek chunk ka thoda hissa agle chunk mein repeat hota hai, taake context preserve ho.
- **LangChain**: Ek library jo splitting ko manage karne mein madad deti hai, aur text ko **documents** mein convert karti hai jo baad mein language model ke liye useful hote hain.

Yeh approach aapki applications ke lamba text handle karne ki capability ko improve karti hai.


### Separator

**Separator** ka kaam hai text ko todne ke liye ek specific point tay karna. Jab aap text ko split karte hain, to separator ye batata hai ke text kis jagah par break hoga.

#### Example:

Maan lo aapka text hai:

```
"Hello! How are you? I hope you're doing well."
```

Agar aap separator ko question mark (`?`) set karte hain, to aapka text is tarah se split hoga:

1. **Chunk 1**: "Hello! How are you"
2. **Chunk 2**: " I hope you're doing well."

Yahan, question mark (`?`) ko separator banaya gaya hai. Is wajah se text wahan se tod diya gaya.

### Chunk Overlap

**Chunk overlap** ka matlab hai ke aapke chunks ke beech thoda sa text shared hoga. Ye aapko madad karta hai taake context na kho jaye. 

#### Example:

Maan lo aapka text ye hai:

```
"This is a test sentence for overlap."
```

Agar aap chunk size ko **20** aur chunk overlap ko **5** set karte hain, to text is tarah se split hoga:

1. **Chunk 1**: "This is a test sen"
2. **Chunk 2**: "tence for overlap."

Yahan par, "sen" jo chunk 1 ke end mein hai, wo chunk 2 ke shuru mein bhi hai. Is tarah se overlap aapko ye ensure karta hai ke kuch context do chunks mein shared hai.

### Conclusion

- **Separator** se aap text ko specific points par todte hain, jaise `?`, `!`, ya `.`.
- **Chunk overlap** se aap ke chunks mein kuch common text hota hai, jo context ko maintain karta hai.

Agar aapko koi aur confusion hai ya aur examples chahiye, to zaroor bataiye!

### Level 2: Recursive Character Text Splitting

**Recursive Character Text Splitting** ka matlab hai ke hum text ko uski structure ya form ke hisaab se taqseem karte hain. Iska maqsad yeh hai ke hum sirf characters ki ginti nahi karte, balki text ke **separators** ka istemal karte hain. 

### Kya Hain Separators?
Separators woh cheezein hain jin par hum text ko todte hain. LangChain mein kuch common separators hain:

1. **Double new line (`"\n\n"`)**: Yeh aksar paragraphs ke beech ka farq dikhata hai.
2. **New line (`"\n"`)**: Yeh ek naya line shuru karne ke liye hota hai.
3. **Space (`" "`)**: Yeh alag alag lafzon ko milane mein madad karta hai.
4. **Characters (`""`)**: Yeh individual characters ke liye hai.

### Kyun Yeh Acha Hai?
Pehle level mein, hum sirf ek fixed number ke characters ka istemal karte the, jisse kabhi kabhi text ka matlab khatam ho jata tha. Lekin is method se, hum text ko **structured** tareeqe se split karte hain, jo zyada understandable hota hai.

### .create_documents
`create_documents` method ka istemal `RecursiveCharacterTextSplitter` ya kisi aur text splitter ke saath hota hai taake text ko chunks mein convert kiya ja sake, jo ke downstream tasks ke liye use kiye ja sakte hain. Is method ka kaam hai text ko alag-alag documents mein split karna, jisse aapko choti choti units milein jo semantic meaning rakhti hain.

### `create_documents` Method Ka Kaam

1. **Input**: Is method ko ek list of strings di jati hai. Yeh strings woh text hote hain jinhe aap split karna chahte hain. 
   
2. **Output**: Yeh method `Document` objects ki ek list return karta hai. Har `Document` object mein `page_content` hota hai, jo chunked text ko represent karta hai.

3. **Chunki Gai Text**: Jab aap `create_documents` call karte hain, to yeh text ko defined separators ke zariye split karta hai. Phir yeh ensure karta hai ke har chunk ki size `chunk_size` parameter se kam ho.


### 1. **Document Specific Splitting Kya Hai?**
Document specific splitting ka matlab hai alag-alag tarah ke documents (jaise Markdown, Python code, JavaScript, PDFs) ko unki khasiyat ke mutabiq chhote hisson (chunks) mein todna. Har type ka document alag tareeqe se split hota hai.

### 2. **Markdown**
Markdown ek format hai jo headings aur sections ko define karta hai. Isme alag alag sections ko split karne ke liye special characters use hote hain. 

**Example:**
```python
from langchain.text_splitter import MarkdownTextSplitter

markdown_text = """
# Fun in California
## Driving
Try driving on the 1 down to San Diego
### Food
Make sure to eat a burrito while you're there
## Hiking
Go to Yosemite
"""

splitter = MarkdownTextSplitter(chunk_size=40, chunk_overlap=0)
documents = splitter.create_documents([markdown_text])
```
Is example mein, Markdown text ko headings aur paragraphs ke hisaab se chhote hisson mein baanta gaya hai. Lekin kabhi kabhi, akele shabd bhi alag chunk ban sakte hain (jaise "there").

### 3. **Python Code**
Python code mein classes, functions aur statements hote hain. Har type ke code ka alag structure hota hai, isliye isay alag tareeqe se split karna padta hai.

**Example:**
```python
from langchain.text_splitter import PythonCodeTextSplitter

python_text = """
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("John", 36)

for i in range(10):
    print(i)
"""

python_splitter = PythonCodeTextSplitter(chunk_size=100, chunk_overlap=0)
documents = python_splitter.create_documents([python_text])
```
Is example mein, Python code ko classes aur functions ke hisaab se split kiya gaya hai.

### 4. **JavaScript**
JavaScript ka bhi kuch aisa hi hai, jahan functions aur variables ko alag alag identify karke split kiya jata hai.

**Example:**
```python
from langchain.text_splitter import RecursiveCharacterTextSplitter, Language

javascript_text = """
// Function is called, the return value will end up in x
let x = myFunction(4, 3);

function myFunction(a, b) {
  return a * b;
}
"""

js_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.JS, chunk_size=65, chunk_overlap=0
)
documents = js_splitter.create_documents([javascript_text])
```
Isme bhi code ko functions aur variables ke hisaab se alag kiya gaya hai.

### Semantic chunking

Semantic chunking ka yeh process yeh kehta hai ke jab hum text ko chote hisson (chunks) mein divide karte hain, toh hum normally ek fixed size chunk use karte hain, jaise ke 500 words, etc. Lekin yeh approach weird lagti hai, kyunki yeh text ke asal meaning ko nahi dekhti. Isme sirf quantity dekhte hain, quality ya meaning nahi.

Is liye author ka idea yeh hai ke hum text ko samajhne ki koshish karein aur jahan ek sentence ka meaning agle sentence se bilkul alag ho jaye, wahan hum ek naya chunk banayein. Is technique ko semantic chunking kaha ja raha hai, kyunki hum content ka meaning (semantics) dekh rahe hain.

### Semantic Chunking Method:

1. **Embeddings Use Karna**: Sabse pehle, hum har sentence ka meaning embedding ke zariye capture karte hain. Embedding ka matlab hota hai ek number representation jo sentence ke meaning ko samjhay. OpenAI ka embedding use kiya gaya.

2. **Sentences Ko Combine Karna**: Short sentences aksar itna meaning nahi rakhte jitna chote hisson ko mila kar rakha ja sake. Is liye, har sentence ke aage aur piche waale sentences ko bhi le kar ek "combined sentence" banaya gaya.

3. **Cosine Similarity**: Har combined sentence ke embeddings ke darmiyan distance measure kiya gaya. Cosine similarity ka use kiya gaya jo batata hai ke do embeddings kitne same ya different hain.

4. **Breakpoints Dhoondhna**: Jab do sentences ke darmiyan distance zyada ho jaye, matlab unka meaning bilkul different ho, toh us jagah ko ek breakpoint samjha gaya aur naya chunk shuru kiya gaya. Yahan distance ki limit 95th percentile rakhi gayi, yani sirf wo jagah chunni gayi jahan distance kaafi zyada tha.

5. **Chunks Banane Ka Tareeqa**: Jab breakpoints mil gaye, toh text ko in jagah par divide kar diya gaya, jahan distance threshold se zyada tha. Har chunk ka meaning ek dosray se related hota hai.

### Result:

Is tareeqe se text ko chunk kiya gaya. Yeh technique aapko text ko better samajhne mein madad deti hai, kyunki yeh content ke meaning ko dekhti hai, sirf text ki length ko nahi.

Aakhir mein, yeh abhi experimental method hai aur ise aur improve kiya ja sakta hai.

### Semantic chunking

Matlab, jab ek hissa doosre hisson se alag ya mukhtalif ho, wahan hum text ko divide kar dete hain.

3. **Chunks banana**: Is tarah se, jab hume lagta hai ke do hisson ka matlab mukhtalif hai, hum unko alag alag chunks mein daal dete hain. Yeh chunks kaafi meaningful hote hain, kyunki hum unko meaning ke hisaab se divide karte hain, na ke sirf length ke hisaab se.

Is tarah, har hissa (chunk) ka matlab ek dosre se related hota hai.

Semantic chunking ka matlab hai text ko aisay hisson mein taqseem karna jo ke ma'ni aur context ko samajhne mein madadgar hon. Jab aap `breakpoint_threshold_type="percentile"` ka istemal karte hain, toh ye text ko ma'ni ke buniyad par taqseem karta hai, matlab jab text mein soch ya meaning ka farq hota hai, tab wahan pe hum break lagate hain.
